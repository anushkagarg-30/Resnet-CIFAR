# Residual Network (ResNet) Architecture

## Details and Instructions

Recall that a residual network (ResNet) architecture is any convolutional network with skipped connections.
Here is the architecture of ResNet-18:

![ResNet-18](resnet18.png)

The key component in ResNet models is a residual block that implements:

$$\text{ReLU}(S(x) + F(x))$$

where \( S(x) \) refers to the skipped connection and \( F(x) \) is a block that implements:

```
conv -> BN -> relu -> conv -> BN
```

Here, **“BN”** stands for batch normalization.  
Chaining such blocks layer by layer gives a deep ResNet.

---

## **Hyperparameters (Design Variables)**

Hyperparameters in such architectures include:

- \( C_i \), the number of channels in the \( i \)-th layer.  
- \( F_i \), the filter size in the \( i \)-th layer.  
- \( K_i \), the kernel size in the \( i \)-th skip connection.  
- \( P \), the pool size in the average pool layer.  
- etc.

# Data Augmentation and Preprocessing for Training Images

The below code creates a pipeline of image transformations that will be applied to our training dataset. Each image will go through these steps in order:

## 1. Random Cropping
* First, each 32×32 image is padded with 4 pixels on all sides
* Then, a random 32×32 crop is taken from this padded image
* This technique introduces slight position variations, helping our model become more robust to object placement

## 2. Random Horizontal Flipping
* Each image has a 50% chance of being flipped horizontally
* This is a simple but effective way to double our effective dataset size
* Helps the model learn that left-right orientation doesn't change an object's category

## 3. Conversion to PyTorch Tensor
* Converts PIL image objects to PyTorch tensors
* Automatically scales pixel values from the range [0, 255] to [0, 1]
* Arranges the tensor in the format [channels, height, width] that PyTorch expects

## 4. Normalization
* Standardizes each color channel using CIFAR-10's statistics:
   * Mean values: (0.4914, 0.4822, 0.4465) for R, G, B channels
   * Standard deviation values: (0.2023, 0.1994, 0.2010)
* This centers the data around zero and gives it unit variance
* Normalization helps with faster and more stable training

These transformations work together to create varied training examples from our original dataset, which helps prevent overfitting and improves the model's ability to generalize to new images.

````python
transform_train = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])
````


# Custom ResNet Model Architecture

Below is the code where we have implemented a customized ResNet architecture for the CIFAR-10 classification task. The implementation focuses on creating an efficient model with less than 5 million parameters while maintaining high accuracy. Let's break down the components in detail.

![ResNet-18](ImprovedResNet18.png)

## Basic Residual Block (`BasicBlock` Class)

The `BasicBlock` class is the fundamental building block of our ResNet architecture. Each block contains:

### Components:

1. **First Convolution Block**:
   - `self.conv1`: A 3×3 convolutional layer that takes `in_planes` input channels and produces `planes` output channels
   - `stride` parameter controls downsampling (stride=2 halves the spatial dimensions)
   - `padding=1` maintains spatial dimensions when stride=1
   - `bias=False` because we're using batch normalization after convolution

2. **First Normalization and Activation**:
   - `self.bn1`: Batch normalization for stabilizing training and improving convergence
   - `self.dropout1`: Dropout layer with 20% drop probability to prevent overfitting

3. **Second Convolution Block**:
   - `self.conv2`: Another 3×3 convolution that maintains the number of channels (`planes`)
   - Always uses `stride=1` to maintain spatial dimensions
   - Similar padding and bias settings as the first convolution

4. **Second Normalization**:
   - `self.bn2`: Batch normalization after the second convolution

5. **Skip Connection (Shortcut)**:
   - `self.shortcut`: Identity mapping by default
   - When dimensions change (due to stride>1 or channel count changes), a 1×1 convolution with batch normalization is used to match dimensions

### Forward Pass:

The forward pass implements the key ResNet formula: `ReLU(S(x) + F(x))` where:
- `F(x)` is the transformation through the convolutional path
- `S(x)` is the shortcut path
- Their sum is then passed through ReLU activation

Key operations:
1. Input `x` is preserved as `identity` for the skip connection
2. The main path applies convolution → batch norm → ReLU → dropout → convolution → batch norm
3. The result is added to the shortcut connection (which may include transformations to match dimensions)
4. ReLU activation is applied to the sum

## ImprovedResNet Main Architecture

The `ImprovedResNet` class builds the full network by stacking multiple residual blocks:

### Components:

1. **Initial Layer**:
   - `self.conv1`: Initial 3×3 convolution that takes RGB images (3 channels) and produces 32 feature maps
   - `self.bn1`: Batch normalization
   - `self.dropout`: Dropout with 25% probability

2. **Residual Layers**:
   - Four layers of stacked residual blocks: `layer1`, `layer2`, `layer3`, and `layer4`
   - Each created by the `_make_layer` method
   - Progressive increase in feature channels: 64 → 128 → 192 → 256
   - Downsampling occurs in layers 2, 3, and 4 (stride=2)

3. **Final Classification Head**:
   - `self.bn_final`: Batch normalization on the flattened features
   - `self.dropout_final`: Higher dropout (30%) before the final classification
   - `self.linear`: Linear layer for classification with 10 outputs (one per CIFAR-10 class)

4. **Weight Initialization**:
   - Custom weight initialization through `_initialize_weights`

### Layer Creation:

This method:
1. Creates a list of strides - first block may downsample (stride>1), rest maintain dimensions (stride=1)
2. Builds a list of residual blocks
3. Updates the running count of input channels (`self.in_planes`)
4. Returns blocks as a sequential module

### Weight Initialization:

Custom initialization for better training:
1. **Convolutional layers**: Kaiming initialization optimized for ReLU
2. **Batch norm layers**: Weight initialized to 1.0 (identity) with small positive bias
3. **Linear layers**: Xavier uniform initialization for weights, zeros for bias

### Forward Pass:

The data flow in forward pass:
1. Initial convolution, batch norm, ReLU, and dropout
2. Processing through the four residual layers
3. Global average pooling (4×4 kernel) to reduce spatial dimensions
4. Flattening the features with `.view()`
5. Final batch normalization, dropout, and classification


These changes ensure that the mode stays under the 5 million parameter limit.



````python
class BasicBlock(nn.Module):
    expansion = 1
    def __init__(self, in_planes, planes, stride=1):
        super(BasicBlock, self).__init__()
        self.conv1 = nn.Conv2d(
            in_planes, planes, kernel_size=3, stride=stride, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(planes)
        self.dropout1 = nn.Dropout(0.2)  
        
        self.conv2 = nn.Conv2d(planes, planes, kernel_size=3,
                              stride=1, padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(planes)
        self.shortcut = nn.Sequential()
        if stride != 1 or in_planes != self.expansion*planes:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_planes, self.expansion*planes,
                         kernel_size=1, stride=stride, bias=False),
                nn.BatchNorm2d(self.expansion*planes)
            )
    
    def forward(self, x):
        identity = x
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.dropout1(out)  
        out = self.bn2(self.conv2(out))
        out += self.shortcut(identity)
        out = F.relu(out)
        return out

class ImprovedResNet(nn.Module):
    def __init__(self, block, num_blocks, num_classes=10):
        super(ImprovedResNet, self).__init__()
        self.in_planes = 32
        
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3,
                              stride=1, padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(32)
        self.dropout = nn.Dropout(0.25)  
        
        self.layer1 = self._make_layer(block, 64, num_blocks[0], stride=1)
        self.layer2 = self._make_layer(block, 128, num_blocks[1], stride=2)
        self.layer3 = self._make_layer(block, 192, num_blocks[2], stride=2)
        self.layer4 = self._make_layer(block, 256, num_blocks[3], stride=2)
        
        
        self.bn_final = nn.BatchNorm1d(256*block.expansion)
        self.dropout_final = nn.Dropout(0.3)  
        self.linear = nn.Linear(256*block.expansion, num_classes)
        
        
        self._initialize_weights()
        
    def _make_layer(self, block, planes, num_blocks, stride):
        strides = [stride] + [1]*(num_blocks-1)
        layers = []
        for stride in strides:
            layers.append(block(self.in_planes, planes, stride))
            self.in_planes = planes * block.expansion
        return nn.Sequential(*layers)
    
    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, (nn.BatchNorm2d, nn.BatchNorm1d)):
               
                nn.init.constant_(m.weight, 1.0)
                nn.init.constant_(m.bias, 0.01)  # Small positive bias
            elif isinstance(m, nn.Linear):
                
                nn.init.xavier_uniform_(m.weight)
                nn.init.constant_(m.bias, 0)
                
    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.dropout(out)  
        
        out = self.layer1(out)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        
        out = F.avg_pool2d(out, 4)  
        out = out.view(out.size(0), -1)
        
        
        out = self.bn_final(out)
        out = self.dropout_final(out)
        out = self.linear(out)
        return out
````

````python
def ResNet18(num_classes=10):
    return ImprovedResNet(BasicBlock, [2, 2, 2, 2], num_classes)
````

## Training and Optimization

For training our ResNet18 model, we implemented a  training process which tries to maximize accuracy while preventing overfitting. Our training approach ran for 120 epochs.

### Optimization Setup

- **Optimizer**: We used SGD with the following parameters:
  * Learning rate: 0.01 
  * Momentum: 0.9 which helps to accelerate the convergence.
  * Weight decay: 0.0005 i.e L2 regularization to prevent overfitting
  * Nesterov momentum: Enabled which provides better convergence properties

- **Learning Rate Scheduling**: We implemented OneCycleLR scheduler with these parameters:
  * Maximum learning rate: 0.1
  * Percentage start: 30% learning rate increases for the first 30% of training, then decreases
  * Annealing strategy: Cosine
  * Division factors: Initial div_factor=10.0, final_div_factor=1000.0
  * This schedule allowed the learning rate to first increase from 0.01 to 0.1, then gradually decrease to 0.0001

- **Loss Function**: Cross-entropy loss with label smoothing of 0.1
  * Label smoothing prevented the model from becoming overconfident and improved generalization

### Training Loop Implementation

For each epoch, our training process:

1. **Training Phase**:
   * Set model to training mode (`model.train()`)
   * Used mixed precision training with GradScaler when GPU was available
   * Applied gradient clipping with max_norm=1.0 to prevent exploding gradients
   * Updated learning rate after each batch with the scheduler
   * Tracked training loss and accuracy

2. **Validation Phase**:
   * Set model to evaluation mode (`model.eval()`)
   * Disabled gradient calculation for efficiency
   * Computed validation loss and accuracy
   * Stored metrics for later analysis

3. **Model Checkpointing**:
   * Saved the model whenever validation accuracy improved
   * Implemented early stopping with patience=10 epochs
   * Stored optimizer state along with model weights for possible resumption



````python
epochs=120
train_loss_history = []
val_loss_history = []
train_acc_history = []
val_acc_history = []


optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9, 
                    weight_decay=0.0005, nesterov=True)

scheduler = torch.optim.lr_scheduler.OneCycleLR(
    optimizer,
    max_lr=0.1,  
    epochs=epochs,
    steps_per_epoch=len(trainDataLoader),
    pct_start=0.3, 
    anneal_strategy='cos',
    div_factor=10.0,  
    final_div_factor=1000.0  
)

criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

scaler = torch.amp.GradScaler('cuda') if torch.cuda.is_available() else None

best_acc = 0.0
patience = 10  
early_stop_counter = 0


for epoch in range(epochs):
   
    model.train()
    train_loss = 0.0
    correct = 0
    total = 0
    
    for inputs, targets in trainDataLoader:
        inputs, targets = inputs.to('cuda'), targets.to('cuda')
        optimizer.zero_grad()
        
        
        if scaler is not None:
            with torch.amp.autocast('cuda'):
                outputs = model(inputs)
                loss = criterion(outputs, targets)
            
            
            scaler.scale(loss).backward()
            
            
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            
            scaler.step(optimizer)
            scaler.update()
        else:
            
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            
            
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            optimizer.step()
        
        
        scheduler.step()
        
        train_loss += loss.item()
        _, predicted = outputs.max(1)
        total += targets.size(0)
        correct += predicted.eq(targets).sum().item()
    
    train_loss = train_loss / len(trainDataLoader)
    train_acc = 100. * correct / total
    
    
    model.eval()
    val_loss = 0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for inputs, targets in valDataLoader:
            inputs, targets = inputs.to('cuda'), targets.to('cuda')
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            val_loss += loss.item()
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()
    
    val_loss = val_loss / len(valDataLoader)
    val_acc = 100. * correct / total
    
    # Save metrics to history lists
    train_loss_history.append(train_loss)
    val_loss_history.append(val_loss)
    train_acc_history.append(train_acc)
    val_acc_history.append(val_acc)
    
    # Save best model
    if val_acc > best_acc:
        best_acc = val_acc
        best_model_wts = model.state_dict().copy()
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'accuracy': val_acc,
        }, 'best_model_generalization.pth')
        early_stop_counter = 0
    else:
        early_stop_counter += 1
    
    
    print(f'Epoch: {epoch+1}/120 | Train Loss: {train_loss:.3f} | Train Acc: {train_acc:.3f}% | Val Loss: {val_loss:.3f} | Val Acc: {val_acc:.3f}%')


model.load_state_dict(best_model_wts)
````

````python
def plot_confusion_matrix(model, data_loader, class_names, device):
    
    model.eval()
    
   
    y_true = []
    y_pred = []
    
    
    with torch.no_grad():
        for inputs, targets in data_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            
            
            y_true.extend(targets.cpu().numpy())
            y_pred.extend(predicted.cpu().numpy())
    
    
    cm = confusion_matrix(y_true, y_pred)
    
   
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.title('Normalized Confusion Matrix')
    plt.tight_layout()
    plt.show()
    
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.title('Confusion Matrix (Counts)')
    plt.tight_layout()
    plt.show()
    
    return cm, cm_normalized


class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
               'dog', 'frog', 'horse', 'ship', 'truck']

plot_confusion_matrix(model, valDataLoader, class_names, device)
````

````python
class FeatureMapVisualizer:
    def __init__(self, model):
        self.model = model
        self.hooks = []
        self.feature_maps = {}
        
    def _get_activation(self, name):
        def hook(model, input, output):
            self.feature_maps[name] = output.detach()
        return hook
    
    def register_hooks(self, layers_to_visualize):
      
        for name, layer in layers_to_visualize.items():
            hook = layer.register_forward_hook(self._get_activation(name))
            self.hooks.append(hook)
    
    def remove_hooks(self):
        
        for hook in self.hooks:
            hook.remove()
        self.hooks = []
    
    def visualize_feature_maps(self, input_image, layer_name, num_features=16, figsize=(15, 8)):
        
        if len(input_image.shape) == 3:
            input_image = input_image.unsqueeze(0)
        
        
        self.model.eval()
        device = next(self.model.parameters()).device
        input_image = input_image.to(device)
        
        # Forward pass
        with torch.no_grad():
            _ = self.model(input_image)
        
       
        if layer_name not in self.feature_maps:
            raise ValueError(f"Layer {layer_name} not found. Available layers: {list(self.feature_maps.keys())}")
        
        feature_map = self.feature_maps[layer_name][0]  
        feature_map = feature_map.cpu().numpy()
        
        
        n_features = min(num_features, feature_map.shape[0])
        grid_size = int(np.ceil(np.sqrt(n_features)))
        
        
        plt.figure(figsize=figsize)
        
        for i in range(n_features):
            plt.subplot(grid_size, grid_size, i + 1)
            
            
            plt.imshow(feature_map[i], cmap='viridis')
            plt.axis('off')
            plt.title(f'Filter {i}')
        
        plt.suptitle(f'Feature Maps for Layer: {layer_name}', fontsize=16)
        plt.tight_layout()
        plt.subplots_adjust(top=0.9)  # Adjust to make room for the suptitle
        plt.show()
        
    def visualize_multiple_layers(self, input_image, figsize=(20, 15)):
        """
        Visualize feature maps from all registered layers
        
        Args:
            input_image: Tensor of shape [C, H, W] or [1, C, H, W]
            figsize: Size of the figure
        """
        
        if len(input_image.shape) == 3:
            input_image = input_image.unsqueeze(0)
        
        
        self.model.eval()
        device = next(self.model.parameters()).device
        input_image = input_image.to(device)
        
       
        with torch.no_grad():
            _ = self.model(input_image)
        
        
        n_layers = len(self.feature_maps)
        
        
        fig, axes = plt.subplots(n_layers, 8, figsize=figsize)
        fig.subplots_adjust(hspace=0.4, wspace=0.4)
        
        for idx, (layer_name, feature_map) in enumerate(self.feature_maps.items()):
            feature_map = feature_map[0].cpu().numpy()  
            
           
            if idx == 0:
                
                orig_img = input_image[0].cpu().numpy()
                
                orig_img = np.transpose(orig_img, (1, 2, 0))
                
                mean = np.array([0.4914, 0.4822, 0.4465])
                std = np.array([0.2023, 0.1994, 0.2010])
                orig_img = std * orig_img + mean
                orig_img = np.clip(orig_img, 0, 1)
                
                plt.figure(figsize=(4, 4))
                plt.imshow(orig_img)
                plt.title('Original Image')
                plt.axis('off')
                plt.show()
            
            
            n_features = min(8, feature_map.shape[0])
            
            for i in range(n_features):
                if n_layers == 1:
                    ax = axes[i]
                else:
                    ax = axes[idx, i]
                
                im = ax.imshow(feature_map[i], cmap='viridis')
                ax.set_title(f'Layer {layer_name}\nFilter {i}')
                ax.axis('off')
            
            # Turn off any unused subplots
            for i in range(n_features, 8):
                if n_layers == 1:
                    axes[i].axis('off')
                else:
                    axes[idx, i].axis('off')
        
        plt.tight_layout()
        plt.show()


print("Generating feature map visualizations...")

sample_batch = next(iter(valDataLoader))
sample_image = sample_batch[0][0]  # First image in batch


layers_to_visualize = {
    'initial': model.conv1,
    'layer1_start': model.layer1[0].conv1,
    'layer1_end': model.layer1[-1].conv2,
    'layer2_middle': model.layer2[0].conv1,
    'layer3_middle': model.layer3[0].conv1,
    'layer4_end': model.layer4[-1].conv2
}


visualizer = FeatureMapVisualizer(model)
visualizer.register_hooks(layers_to_visualize)

try:
   
    print("Visualizing individual layer feature maps...")
    for layer_name in layers_to_visualize.keys():
        print(f"Visualizing {layer_name}...")
        visualizer.visualize_feature_maps(sample_image, layer_name)
    
    
    print("Visualizing all layers together...")
    visualizer.visualize_multiple_layers(sample_image)
finally:
   
    visualizer.remove_hooks()

print("Visualization complete!")
````

## Prediction Process

For the final step of our project, we processed the unlabeled test data and generated predictions for the Kaggle competition. Here's how we approached it:

### Preparing the Test Data
We started by unpickling the CIFAR-10 test data file, which gave us 10,000 unlabeled images and their IDs. The data initially came in a flattened format (10000×3072), so we needed to reshape it into the proper image format (10000×3×32×32) that our model expects.

Next, we applied the same normalization steps we used during training:
- Divided pixel values by 255 to scale them to [0,1]
- Subtracted the CIFAR-10 mean values [0.4914, 0.4822, 0.4465]
- Divided by the standard deviations [0.2023, 0.1994, 0.2010]

This preprocessing ensures that the test data matches the distribution our model was trained on.

### Making Predictions
With the data ready, we:
1. Created a DataLoader to batch the test images (batch size=64)
2. Set our model to evaluation mode to disable dropout
3. Ran each batch through the model and collected the predicted class labels
4. Created a submission DataFrame with each image ID and its predicted label
5. Saved the results to a CSV file for Kaggle submission

### Visualizing Results
To double-check our predictions, we created a visualization grid showing some test images along with their predicted classes. This helped us verify that the model was making reasonable classifications before submitting.

This visual check gave us confidence that our model was performing well on the test set. Based on our validation accuracy of 94.29%, we're optimistic about how our submission will rank in the competition.

````python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict


test_data = unpickle('cifar_test_nolabel.pkl')


x_test_nolabel = test_data.get(b'data')
ids = test_data.get(b'ids')
print(f"Test data shape: {np.array(x_test_nolabel).shape}")
print(f"Number of test samples: {len(ids)}")


if len(np.array(x_test_nolabel).shape) == 2:
    
    x_test_reshaped = np.array(x_test_nolabel).reshape(-1, 3, 32, 32)
    
    x_test_viz = np.transpose(x_test_reshaped, (0, 2, 3, 1))
else:
    
    x_test_viz = np.array(x_test_nolabel)
    x_test_reshaped = np.transpose(x_test_viz, (0, 3, 1, 2))


x_test_tensor = torch.tensor(x_test_reshaped, dtype=torch.float32) / 255.0


mean = torch.tensor([0.4914, 0.4822, 0.4465]).view(3, 1, 1)
std = torch.tensor([0.2023, 0.1994, 0.2010]).view(3, 1, 1)
x_test_tensor = (x_test_tensor - mean) / std


test_dataset = TensorDataset(x_test_tensor)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)




model = model.to(device)
model.eval()


all_predictions = []
with torch.no_grad():
    for batch in test_loader:
        images = batch[0].to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        all_predictions.extend(predicted.cpu().numpy())


results = pd.DataFrame({
    'ID': ids,
    'Label': all_predictions
})


results.to_csv('submission.csv', index=False)
print(f"Predictions saved to submission.csv")


class_names = {
    0: 'airplane', 1: 'automobile', 2: 'bird', 3: 'cat',
    4: 'deer', 5: 'dog', 6: 'frog', 7: 'horse', 8: 'ship', 9: 'truck'
}


def visualize_predictions(data, predictions_df, num_images=15):
    
    num_to_show = min(num_images, len(data))

    
    rows = (num_to_show + 4) // 5
    cols = min(5, num_to_show)

    plt.figure(figsize=(3*cols, 3*rows))

    for i in range(num_to_show):
        
        img = data[i]

       
        pred_idx = predictions_df.iloc[i]['Label']
        pred_name = class_names[pred_idx]
        img_id = predictions_df.iloc[i]['ID']

        
        plt.subplot(rows, cols, i+1)
        plt.imshow(img)
        plt.title(f"ID: {img_id}\nPred: {pred_name}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()


visualize_predictions(x_test_viz, results, num_images=150)
````
